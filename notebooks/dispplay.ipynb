```python
# Charger les données depuis un fichier texte
import matplotlib.pyplot as plt
import pandas as pd
import re


file_path = "../logs/graph_log.txt"

# Lire le contenu du fichier
with open(file_path, 'r') as file:
    log_data = file.read()

# Extraire les métriques à l'aide des expressions régulières
pattern = r"Epoch (\d+) .* val acc (\d+\.\d+)%, .* val precision (\d+\.\d+)%, val recall (\d+\.\d+)%"
matches = re.findall(pattern, log_data)

# Créer un DataFrame à partir des données extraites
data = pd.DataFrame(matches, columns=["Epoch", "Accuracy", "Precision", "Recall"])
data = data.astype({"Epoch": int, "Accuracy": float, "Precision": float, "Recall": float})

# Affichage graphique des métriques
plt.figure(figsize=(12, 6))
plt.plot(data["Epoch"], data["Accuracy"], label="Accuracy")
plt.plot(data["Epoch"], data["Precision"], label="Precision")
plt.plot(data["Epoch"], data["Recall"], label="Recall")
plt.xlabel("Epoch")
plt.ylabel("Percentage")
plt.title("Performance Metrics Over Epochs")
plt.legend()
plt.grid()
plt.show()

```
